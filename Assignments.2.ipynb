```python
list = [10,20,30,40,50]
print("original list:",list)
## Add the element to the end of list
list.append(60)
print("after append(60):",list)
##Extends the list by appending all the items from an iterable
list.extend([70,80])
print(" after extend([70,80]):",list)   
##Inserts an element at a specified position
list.insert(1,15)
print("after insert(1,15):",list)
##Removes the first occurrence of a specified value
list.remove(50)
print("after remove(50):",list)
##Removes and returns the element at a specified index (or the last element if no index is given)
list.pop()
print("after pop():",list)
##Removes all elements from the list
list.clear()
print("after clear():",list)
list = [10, 20, 30, 40, 50]
print("original list:",list)
##Returns the index of the first occurrence of a value
index_40 = list.index(40)
print("index of (40):",index_40)
## Returns the number of times a specified value appears in the list
count_20 = list.count(20)
print("count of 20 in the list:",count_20)
##Sorts the list in ascending order
list.sort()
print("after sort():",list)
##Reverses the order of elements in the list
list.reverse()
print("after reverse():",list)
```

```python
# Take input from user
num = int(input("Enter a number: "))
# Check conditions
if num > 0:
    print("The number is positive.")
    if num % 2 == 0:
        print("It is also even.")
    else:
        print("It is odd.")
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```

```python
import pandas as pd
import numpy as np
# Create a DataFrame manually with 10 rows
data = {
    'Name': ['Tanay', 'Priya', 'Tanishk', 'Sneha', 'Ravi', 'priti', 'Neha', 'Rutuja', 'Riya', 'samarth'],
    'Age': [25, 30, 22, 28, 26, 25, 27, 32, 29, 22],
    'City': ['Mumbai', 'Pune', 'Delhi', 'Chennai', 'Kolkata', 'Mumbai', 'Delhi', 'Pune', 'chennai', 'Delhi'],
    'Salary': [50000, 60000, np.nan , 58000, 62000, 50000, 55000,np.nan, 75000, 58000]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
#  Display info and statistics
print("\nDataFrame Info:")
print(df.info())
print("\nDescriptive Statistics:")
print(df.describe())
#  Select 'Name' and 'City' columns
print("\nSelected Columns (Name & City):\n", df[['Name', 'City']])
#  Drop 'City' column
df_dropped = df.drop(columns=['City'])
print("\nAfter Dropping 'City' Column:\n", df_dropped)
#  Fill missing Salary values with mean
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
print("\nAfter Filling Missing Salary Values with Mean:\n", df)
# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nAfter Removing Duplicate Rows:\n", df_no_duplicates)
```
