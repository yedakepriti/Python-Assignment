```python
import pandas as pd
df = pd.read_csv('/content/sample_data/mnist_test.csv')

# Step 3: Identify numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64'])
print("Numerical Columns:")
print(numerical_columns.columns)

# Step 4: Calculate statistical measures
mean_values = numerical_columns.mean()
median_values = numerical_columns.median()
mode_values = numerical_columns.mode().iloc[0]
std_values = numerical_columns.std()

# Step 5: Display results
print("\nMean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
print("\nStandard Deviation:\n", std_values)
```
