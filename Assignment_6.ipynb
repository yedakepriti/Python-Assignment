```python
import pandas as pd
import numpy as np
from scipy.stats import t, norm

# Sample data (millions of characters)
data = [1.13, 1.55, 1.43, 0.92, 1.25,
        1.36, 1.32, 0.85, 1.07, 1.48,
        1.20, 1.33, 1.18, 1.22, 1.29]

n = len(data)
mean = np.mean(data)
sample_sd = np.std(data, ddof=1)

print("Sample Size =", n)
print("Sample Mean =", round(mean, 4))
print("Sample Standard Deviation =", round(sample_sd, 4))

# Part A: 99% CI using sample standard deviation (t-distribution)
confidence = 0.99
alpha = 1 - confidence
df = n - 1

t_critical = t.ppf(1 - alpha/2, df)
margin_t = t_critical * sample_sd / np.sqrt(n)

lower_t = mean - margin_t
upper_t = mean + margin_t

print("\n99% Confidence Interval using t-distribution")
print("Lower Limit =", round(lower_t, 4))
print("Upper Limit =", round(upper_t, 4))

# Part B: 99% CI using known population standard deviation
sigma = 0.2

z_critical = norm.ppf(1 - alpha/2)
margin_z = z_critical * sigma / np.sqrt(n)

lower_z = mean - margin_z
upper_z = mean + margin_z

print("\n99% Confidence Interval using z-distribution")
print("Lower Limit =", round(lower_z, 4))
print("Upper Limit =", round(upper_z, 4))
```

```python
import pandas as pd
import numpy as np
import math
from scipy.stats import norm

# Given data
x_bar = 3050
mu = 1000 + 5 * 600
sigma = 5 * 25
n = 25
alpha = 0.05

# Z-test statistic
z = (x_bar - mu) / (sigma / math.sqrt(n))

# Critical value (right-tailed test)
z_critical = norm.ppf(1 - alpha)

# p-value
p_value = 1 - norm.cdf(z)

print("Theoretical Mean Cost =", mu)
print("Sample Mean Cost =", x_bar)
print("Z-statistic =", round(z, 4))
print("Critical Value =", round(z_critical, 4))
print("P-value =", round(p_value, 4))

if z > z_critical:
    print("\nDecision: Reject H0")
else:
    print("\nDecision: Fail to Reject H0")
```
